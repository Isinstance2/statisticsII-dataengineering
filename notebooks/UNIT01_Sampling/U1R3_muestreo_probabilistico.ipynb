# Muestreo Sistemático

El muestreo sistemático consiste en seleccionar elementos de una población utilizando un intervalo fijo (**k**) calculado a partir del tamaño de la población y el tamaño de muestra deseado.

Su procedimiento puede resumirse en tres pasos:

1. Calcular el intervalo de selección.
2. Elegir un punto de inicio de manera aleatoria.
3. Seleccionar un elemento cada **k** posiciones hasta completar la muestra.

En Ingeniería de Datos este método resulta especialmente útil cuando se trabaja con registros ordenados cronológicamente, como logs, eventos o transacciones.

# Paso 1

````python
# tamaño de la población
N = len(df)

# tamaño de la muestra
n = 2800

# Intervalo de selección
k = N // n

print(f"Población: {N}")
print(f"Muestra: {n}")
print(f"Intervalo: {k}")


````

Primero calculamos el tamaño de la población (**N**) y definimos el tamaño de muestra (**n**).

Posteriormente obtenemos el intervalo de selección mediante:


Este intervalo indica cada cuántos registros se seleccionará un nuevo elemento.

# Paso 2

````python
inicio = np.random.randint(0, k)
print(f"Inicio aleatorio: {inicio}")
````

El inicio aleatorio evita introducir sesgos derivados de comenzar siempre desde la misma posición del dataset.

Posteriormente todos los registros serán seleccionados utilizando el mismo intervalo.

# Paso 3

````python

# Muestreo sistemático
systematic_sample = df.iloc[inicio::k]

# Ajustar exactamente al tamaño solicitado
systematic_sample = systematic_sample.head(n)

systematic_sample.head()

````

La instrucción:

```python
df.iloc[inicio::k]
```

utiliza el operador **slice** de Python.

Su funcionamiento es:

```python
df.iloc[inicio : fin : paso]
```

En este caso:

- **inicio** → primera posición seleccionada.
- **fin** → hasta el final del DataFrame.
- **paso** → intervalo **k**.

Es decir, Pandas selecciona automáticamente cada **k** registros comenzando desde la posición elegida aleatoriamente.

# Muestreo Estratificado

El muestreo estratificado consiste en dividir la población en grupos llamados **estratos**, donde los individuos de cada grupo comparten una característica en común relacionada con la variable que se desea estudiar.

Una vez creados los estratos, se selecciona una muestra de cada uno de ellos utilizando un método probabilístico, generalmente un muestreo aleatorio simple.

El objetivo es asegurar que todos los grupos importantes de la población estén representados dentro de la muestra final.

> **Aplicación en Ingeniería de Datos**
>
> Este método es especialmente útil cuando un dataset contiene categorías desbalanceadas. Por ejemplo, al entrenar un modelo de Machine Learning, un muestreo estratificado ayuda a mantener la proporción original de las clases y evita que categorías minoritarias desaparezcan durante el proceso de muestreo.

````python
df["cut"].value_counts()
````

````python
(
    df["cut"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
)
````

## Exploración de la Variable de Estratificación

Antes de realizar un muestreo estratificado, es importante conocer cómo está distribuida la variable que utilizaremos para crear los estratos.

En este ejemplo utilizaremos la columna **`cut`**, la cual representa la calidad del corte de cada diamante.

Primero observamos cuántos registros pertenecen a cada categoría.

Podemos observar que la distribución no es uniforme.

- **Ideal** representa la mayor parte del dataset.
- **Fair** contiene una cantidad considerablemente menor de registros.

Si aplicáramos un muestreo completamente aleatorio, existe la posibilidad de que las categorías menos frecuentes queden subrepresentadas dentro de la muestra.

Por esta razón, el muestreo estratificado resulta una excelente alternativa, ya que permite conservar la distribución de la variable **`cut`** dentro de la muestra obtenida.

````python
stratum_sample_size = 1000

stratum_size = (
    df["cut"].value_counts()
    .to_frame(name="population")
)

stratum_size["sample_size"] = (
    stratum_size["population"] / len(df) * stratum_sample_size
).round().astype(int)

stratum_size

````

## Paso 2 - Calcular el tamaño de muestra por estrato

Una vez definido el tamaño total de la muestra, debemos determinar cuántos registros aportará cada estrato.

En un muestreo estratificado proporcional, cada estrato conserva la misma proporción que tiene dentro de la población.

La fórmula utilizada es:


$$
n_h = \frac{N_h}{N}\times n
$$


Donde:

- **N** → Tamaño de la población.
- **Nₕ** → Cantidad de elementos del estrato.
- **n** → Tamaño total de la muestra.
- **nₕ** → Cantidad de elementos que se seleccionarán del estrato.

Observamos que cada estrato aportará una cantidad distinta de registros.

Los grupos con mayor cantidad de observaciones dentro de la población también aportarán una mayor cantidad de elementos a la muestra.

De esta forma se conserva la distribución original de la población.

## Paso 3 - Extraer la muestra de cada estrato

Ahora que conocemos cuántos registros debe aportar cada categoría, realizamos un muestreo aleatorio simple dentro de cada estrato.

Finalmente unimos todas las muestras para obtener la muestra estratificada completa.

````python
stratified_sample = (
    df.groupby("cut", group_keys=False)
      .apply(
          lambda group: group.sample(
              n=stratum_size.loc[group.name, "sample_size"],
              random_state=42
          ),
          include_groups=False
      )
)

stratified_sample.drop(columns="Unnamed: 0", inplace=True)
stratified_sample.head()
````

# Muestreo por Conglomerados

El muestreo por conglomerados consiste en dividir la población en grupos naturales llamados **conglomerados** y seleccionar únicamente algunos de ellos de forma aleatoria.

Una vez seleccionados los conglomerados, se estudian todos los individuos que pertenecen a dichos grupos.

A diferencia del muestreo estratificado, **no se toman elementos de todos los grupos**, sino únicamente de los conglomerados seleccionados.

Esta técnica suele utilizarse cuando resulta costoso acceder a toda la población, ya que permite reducir el tiempo y los recursos necesarios para la recolección de datos.

> **Aplicación en Ingeniería de Datos**
>
> En proyectos de Ingeniería de Datos, un conglomerado puede representar una sucursal, un centro de datos, una región geográfica o un servidor. En lugar de procesar toda la infraestructura, se seleccionan algunos conglomerados y se analizan todos los registros pertenecientes a ellos.

# Muestreo por Estadios Múltiples (Multietápico)

El muestreo por estadios múltiples consiste en seleccionar la muestra en varias etapas consecutivas.

En lugar de acceder directamente a todos los individuos de la población, primero se seleccionan grupos grandes y posteriormente se continúa reduciendo la población mediante nuevas selecciones hasta llegar a las unidades finales de estudio.

Cada etapa puede utilizar uno de los métodos de muestreo probabilístico estudiados anteriormente, como el muestreo aleatorio simple, sistemático, estratificado o por conglomerados.

Este método es ampliamente utilizado cuando la población es muy grande, se encuentra distribuida geográficamente o no existe un listado completo de todos los individuos.

## ¿Por qué utilizar este método?

En muchos proyectos reales resulta imposible acceder directamente a todos los elementos de la población.

Por ejemplo, si una investigación necesita estudiar profesores de un país, primero habría que localizar todas las provincias, luego las ciudades, después las escuelas y finalmente los profesores.

Realizar un muestreo aleatorio simple sobre todos los profesores sería poco práctico, ya que sería necesario disponer de un listado nacional completo y los profesores seleccionados podrían encontrarse muy dispersos geográficamente.

El muestreo por estadios múltiples resuelve este problema reduciendo la población de manera progresiva.

## Representación del proceso

```
País
│
├── Provincia
│      │
│      ├── Municipio
│      │      │
│      │      ├── Escuela
│      │      │      │
│      │      │      └── Profesor
```

Cada nivel representa una nueva etapa del proceso de selección.

## Aplicación en Ingeniería de Datos

En Ingeniería de Datos es común trabajar con datos distribuidos en diferentes niveles de organización.

Por ejemplo:

```
Empresa
│
├── Región
│      │
│      ├── Centro de Datos
│      │      │
│      │      ├── Servidor
│      │      │      │
│      │      │      └── Registros
```

En lugar de procesar todos los registros de toda la infraestructura, es posible seleccionar primero algunas regiones, luego algunos centros de datos, posteriormente algunos servidores y finalmente analizar únicamente los registros pertenecientes a esos servidores.

Este enfoque permite reducir significativamente el costo computacional y el tiempo de procesamiento cuando se trabaja con grandes volúmenes de datos.

> **Nota**
>
> El dataset de diamantes no posee una estructura jerárquica real. Por ello construiremos una jerarquía con fines didácticos para comprender el funcionamiento del método.

````python
selected_colors = np.random.choice(
    df["color"].unique(),
    size=3,
    replace=False
)
````

````python
stage_1 = df[df["color"].isin(selected_colors)]
````

````python
selected_cut = np.random.choice(
    stage_1["cut"].unique(),
    size=3,
    replace=False
)

stage_2 = stage_1[stage_1["cut"].isin(selected_cut)]

final_sample = stage_2.sample(n=1000, random_state=42)

final_sample.drop(columns="Unnamed: 0", inplace=True)
final_sample.head()
````
